```python
#!/usr/bin/python
# coding=utf-8

AUTHOR="齐家乐"

import json
import pandas as pd

types_delete=[]
with open("types_delete.txt","r") as f:
    for i in f:
        types_delete.append(i)


def pat_mat(info):
    for i in types_delete:
        if i==info[0]:
            return [False,""]
    return [True,"test"]
    #在此处添加新模式
    
    
    
    
file="data0.csv"

newTable=pd.DataFrame(columns=["callback_content","vin","label","keyword"])

data=pd.read_csv(file,encoding="utf-8")

for index, row in data.iterrows():
    try:
        records=json.loads(row["callback_content"])
        flag=False
        for rec in records["result"]:
            info=[rec["type"],rec["detail"],rec["other"]]
            ret=pat_mat(info)
            if ret[0]==True:
                newTable.loc[newTable.shape[0]+1]={"callback_content":row["callback_content"],"vin":row["vin"],"label":1,"keyword":ret[1]}
                flag=True
            #print(df)
                break
            else:
                continue
        if flag is False:
            newTable.loc[newTable.shape[0]+1]={"callback_content":row["callback_content"],"vin":row["vin"],"label":1,"keyword":ret[1]}
    except:
        print("except")
        pass
newTable.to_csv("result_"+file,encoding="utf-8",index=False)
            
```
